```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('EsercizioFunghi/mushrooms.csv')
df.head()
```

```python
#Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder

cols=df.columns.to_list()
ordinal_encoder = OrdinalEncoder()
df = pd.DataFrame(ordinal_encoder.fit_transform(df), columns=df.columns.to_list())
#riassegna nomi colonne
df.head()
```

```python
#cross vall tree, svm, random forest
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

X = df.drop('class', axis=1)
y = df['class']

#Decision Tree
dt = DecisionTreeClassifier()
dt_scores = cross_val_score(dt, X, y, cv=5)
#print mean of scores and standard deviation
print('Decision Tree Accuracy and Standard Deviation: ', dt_scores.mean(), dt_scores.std())

#SVM
svm = SVC()
svm_scores = cross_val_score(svm, X, y, cv=5)
print('SVM Accuracy and Standard Deviation: ', svm_scores.mean(), svm_scores.std())

#Random Forest
rf = RandomForestClassifier(n_estimators=40)
rf_scores = cross_val_score(rf, X, y, cv=5)
print('Random Forest Accuracy and Standard Deviation: ', rf_scores.mean(), rf_scores.std())
```

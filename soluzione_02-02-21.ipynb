```
dataset = pd.read_csv('heart_failure_clinical_records_dataset.csv')
Y = dataset['DEATH_EVENT']
X = dataset.drop('DEATH_EVENT', axis=1)
dataset.head()
```

```
X = X.drop(['time'], axis=1)
X
```

```
dis = preprocessing.KBinsDiscretizer(n_bins=6, encode='ordinal', strategy='uniform')
scal = preprocessing.MinMaxScaler()
norm = preprocessing.Normalizer()

prep = ColumnTransformer(transformers=[('discr', dis, ['age','platelets']),
                                      ('scal', scal, ['creatinine_phosphokinase', 'ejection_fraction', 'serum_creatinine', 'serum_sodium'])
                                      ], remainder='passthrough')
```

```
model = DecisionTreeClassifier()

my_pipeline = Pipeline(steps=[('preprocess', prep),
                              ('model', model)
                             ], verbose = True)
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1/3, random_state=0, stratify=dataset['DEATH_EVENT'])
```

```
my_pipeline.fit(X_train, Y_train)
```

```
pred = my_pipeline.predict(X_train)
print('Accuracy Score on TRAIN SET:', accuracy_score(Y_train, pred))
```

```
pred = my_pipeline.predict(X_test)
print('Accuracy Score on TEST SET:', accuracy_score(Y_test, pred))
```

## 5)

```
parameters = {
    'model__criterion': ['gini', 'entropy'],
    'model__max_depth': [5, 7, 10],
    'preprocess__discr__n_bins': [4, 6, 8, 10],
    'preprocess__discr__strategy': ['uniform','quantile'],
}

clf = GridSearchCV(my_pipeline, parameters,  cv=10)
clf.fit(X_train, Y_train)

print("\n\nI MIGLIORI PARAMETRI SONO: {}\n\n".format(clf.best_params_))

my_pipeline.set_params(**clf.best_params_)

# Preprocessing of training data, fit model 
my_pipeline.fit(X_train, Y_train)
pred = my_pipeline.predict(X_train)
print('Accuracy Score on TRAIN SET:', accuracy_score(Y_train, pred))

pred = my_pipeline.predict(X_test)
print('Accuracy Score on TRAIN SET:', accuracy_score(Y_test, pred))
```

## 6)

```
dataset = pd.read_csv('heart_failure_clinical_records_dataset.csv')
Y = dataset['DEATH_EVENT']
X = dataset.drop('DEATH_EVENT', axis=1)
dataset.head()
```

```
X = X.drop(['time'], axis=1)
X
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1/3, random_state=0, stratify=dataset['DEATH_EVENT'])
```

```
dis = preprocessing.KBinsDiscretizer(n_bins=6, encode='ordinal', strategy='uniform')
scal = preprocessing.MinMaxScaler()
norm = preprocessing.Normalizer()

prep = ColumnTransformer(transformers=[('discr', dis, ['age','platelets']),
                                      ('scal', scal, ['creatinine_phosphokinase', 'ejection_fraction', 'serum_creatinine', 'serum_sodium'])
                                      ], remainder='passthrough')
```

```
pca = PCA(n_components=2)
selection = SelectKBest(k=2)

def identity_func(X):
  return X

combined_features = FeatureUnion([("pca", pca), ("select", selection), ('passtrough', FunctionTransformer(identity_func,validate=False))])

# Bundle preprocessing and modeling code in a pipeline
my_pipeline = Pipeline(steps=[('preprocess', prep),
                              ('combined_features', combined_features),
                              ('model', model)
                             ], verbose = True)

# Preprocessing of training data, fit model 
my_pipeline.fit(X_train, Y_train)

# Preprocessing of validation data, get predictions
pred = my_pipeline.predict(X_train)
print('Accuracy Score on TRAIN:', accuracy_score(Y_train, pred))

# Preprocessing of validation data, get predictions
pred = my_pipeline.predict(X_test)
print('Accuracy Score on TEST:', accuracy_score(Y_test, pred))
```
